```python
import random
import numpy as np
import matplotlib.pyplot as plt

#definim els coeficients inicials
a0 = 2 
b0 = 1

#generem 10 punts equidistantment entre 1 i 10
x = np.arange(10)

#assignem cada punt al valor de la recta
yp = a0*x + b0

#afegim soroll gaussià a cada punt
soroll = np.random.normal(0,1, size=10)
y = yp + soroll

#visualitzem les observacions juntament amb la recta inicial
plt.plot(x,yp, label="Recta Original")
plt.scatter(x, y, label="Recta Soroll")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

def f_distances(a,b,x,y):   
    ypred = a*x + b
    dif = ypred - y
    return dif

def get_error(a,b,x,y):
    error = (sum(f_distances(a,b,x,y)**2))/(len(x))
    return error
    
    
import numpy as np
import matplotlib.pyplot as plt

a1 = 2
a2 = -1
a3 = 1
b1 = 2
b2 = 1 
b3 = -2

error1 = get_error(a1,b1,x,y)
error2 = get_error(a2,b2,x,y)
error3 = get_error(a3,b3,x,y)

ypred1 = a1*x + b1
ypred2 = a2*x + b2
ypred3 = a3*x + b3

plt.scatter(x,y,label='y amb soroll')

plt.plot(x,ypred1, label=f'Error 1 = {error1: .2f}')
plt.plot(x,ypred2, label=f'Error 2 = {error2: .2f}')
plt.plot(x,ypred3, label=f'Error 3 = {error3: .2f}')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Aproximació de recta amb error')
plt.show()

import numpy as np
import matplotlib.pyplot as plt


a_vals = np.arange(-5, 5.25, 0.25)
b_vals = np.arange(-5, 5.25, 0.25)

#matriu per emmagatzemar valors
errors = np.zeros((len(a_vals), len(b_vals)))

#calculem els errors i els posem en la matriu
for i, a in enumerate(a_vals):
    for j, b in enumerate(b_vals):
        error = get_error(a, b, x, y)
        errors[i, j] = error

#Visualitzem
plt.imshow(errors, cmap='viridis', origin='lower', extent=[-5, 5, -5, 5], vmin=0, vmax=40)
plt.colorbar(label='Error')


plt.plot([b0], [a0], marker='o', color='red') #El punt sense error (dades generades)

#punts de diferents zones amb diferents mostres d'error per representar en rectes(apartar 3.3)
plt.plot([-2], [-2], marker='x', color='black') #molt error
plt.plot([-2], [3], marker='x', color='black')#poc error
plt.plot([0.5], [2], marker='x', color='black')#molt poc error

plt.xlabel('Valors de b')
plt.ylabel('Valors de a')
plt.title('Errors per diferents valors a i b')


plt.show()

import random
import numpy as np
import math

a = np.random.uniform(-500,500)
b = np.random.uniform(-500,500)
alpha = -0.01

errors = []

for i in range(3000):
    a = a + alpha*(2/10*sum(x*f_distances(a,b,x,y)))
    b = b + alpha*(2/10*sum(f_distances(a,b,x,y)))
    
    error = get_error(a,b,x,y)
    errors.append(error)

#volem representar l'error a escala logarítmica degut a les seves grans proporcions
errorlog=[] #vector amb els logaritmes dels errors
for i in range(3000): 
    errorlogi=math.log(errors[i])
    errorlog.append(errorlogi) 
    
print(f'error final: {error}')
print(f'a estimada: {a} \ta real: {a0}')
print(f'b estimada: {b} \tb real: {b0}')
print('Aquest és el minim error al qual es pot arribar tenint en compte les desviacions de la recta observada')

#visualitzem l'error en funció de les iteracions 
iteracions = np.arange(0,3000)

plt.plot(iteracions,errorlog, label='iteració vs error')
plt.xlabel('iteració')
plt.ylabel('log(error)')  
plt.legend()
plt.show()
    
plt.plot(x,a*x+b,label='y estimada')
plt.plot(x,y,'r--', label='y observada')

plt.xlabel('x')
plt.ylabel('y')    
plt.legend()
plt.show()                    
```
